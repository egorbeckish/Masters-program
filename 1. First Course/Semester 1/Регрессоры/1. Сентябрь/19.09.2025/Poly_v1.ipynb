```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn as skl
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## Полиномиальная регрессия вида $$f(x) = k_1 + k_2x_2 + k_3x_3^2 + ... + k_nx_n^n$$

```python
df = pd.read_csv("files/c-data-1.csv", names=["x^1", "y"])
```

```python
df
```

```python
plt.scatter(df["x^1"], df["y"], edgecolors="black", zorder=3)
plt.grid()
plt.show()
```

```python
degrees = 3
col_degrees = [f"x^{degree}" for degree in range(1, degrees + 1)]
y = df["y"].to_numpy()
```

```python
for degree in range(1, degrees + 1):
	fig, ax = plt.subplots(2, 3, figsize=(20, 8))
	
	curr_degrees = col_degrees[:degree][::-1]
	
	x = df[curr_degrees].to_numpy()
	x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

	model_T = LinearRegression().fit(x, y)
	model_F = LinearRegression(fit_intercept=False).fit(x, y)
	
	model_T_train = LinearRegression().fit(x_train, y_train)
	model_F_train = LinearRegression(fit_intercept=False).fit(x_train, y_train)
	
	model_T_test = LinearRegression().fit(x_test, y_test)
	model_F_test = LinearRegression(fit_intercept=False).fit(x_test, y_test)

	score_T = model_T.score(x, y)
	score_F = model_F.score(x, y)
	coef_T = model_T.coef_
	coef_F = model_F.coef_
	intercept_T = model_T.intercept_
	intercept_F = model_F.intercept_

	score_T_train = model_T_train.score(x_train, y_train)
	score_F_train = model_F_train.score(x_train, y_train)
	coef_T_train = model_T_train.coef_
	coef_F_train = model_F_train.coef_
	intercept_T_train = model_T_train.intercept_
	intercept_F_train = model_F_train.intercept_

	score_T_test = model_T_test.score(x_test, y_test)
	score_F_test = model_F_test.score(x_test, y_test)
	coef_T_test = model_T_test.coef_
	coef_F_test = model_F_test.coef_
	intercept_T_test = model_T_test.intercept_
	intercept_F_test = model_F_test.intercept_

	x, x_train, x_test = np.sort(x, axis=0), np.sort(x_train, axis=0), np.sort(x_test, axis=0)

	for i in range(2):
		for j in range(3):
			ax[i, j].scatter(df["x^1"], df["y"], edgecolors="black", label="Origin data", zorder=3)
			ax[i, j].grid()

	ax[0, 0].set_title("fit_intercept=True")
	ax[0, 1].set_title(f"Polynomial {degree=}")
	
	ax[0, 0].scatter(x[:, -1], model_T.predict(x), label="Regression data", edgecolors="black", zorder=3)
	ax[0, 1].scatter(x_train[:, -1], model_T_train.predict(x_train), label="Training data", edgecolors="black", zorder=3)
	ax[0, 2].scatter(x_test[:, -1], model_T_test.predict(x_test), label="Test data", edgecolors="black", zorder=3)

	ax[1, 0].set_title("fit_intercept=False")
	
	ax[1, 0].scatter(x[:, -1], model_F.predict(x), label="Regression data", edgecolors="black", zorder=3)
	ax[1, 1].scatter(x_train[:, -1], model_F_train.predict(x_train), label="Training data", edgecolors="black", zorder=3)
	ax[1, 2].scatter(x_test[:, -1], model_F_test.predict(x_test), label="Test data", edgecolors="black", zorder=3)
	
	for i in range(2):
		for j in range(3):
			ax[i, j].legend()

	plt.show()
	
	print(
		f"{score_T=}",
		f"{score_F=}",
		f"{coef_T=}",
		f"{coef_F=}",
		f"{intercept_T=}",
		f"{intercept_F=}",
		sep="\n"
	)

	print("\n")

	print(
		f"{score_T_train=}",
		f"{score_F_train=}",
		f"{coef_T_train=}",
		f"{coef_F_train=}",
		f"{intercept_T_train=}",
		f"{intercept_F_train=}",
		sep="\n"
	)

	print("\n")

	print(
		f"{score_T_test=}",
		f"{score_F_test=}",
		f"{coef_T_test=}",
		f"{coef_F_test=}",
		f"{intercept_T_test=}",
		f"{intercept_F_test=}",
		sep="\n"
	)
		
	df[f"x^{degree + 1}"] = df["x^1"] ** (degree + 1)
	df = df[[f"x^{degree + 1}"] + curr_degrees + ["y"]].copy()
```
