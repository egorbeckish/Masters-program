```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso, Ridge, LinearRegression, SGDRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score, GridSearchCV, train_test_split, RandomizedSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from scipy.stats import uniform
```

```python
df = pd.read_csv("files/data-7.csv")
df
```

```python
x, y = df["x"].values.reshape(-1, 1), df["y"].values.reshape(-1, 1)
```

```python
x
```

```python
y
```

```python
poly = PolynomialFeatures(degree=4)

x_poly = poly.fit_transform(x)

x_train_poly, x_test_poly, y_train_poly, y_test_poly = train_test_split(
    x_poly,
    y,
    test_size=.2,
    random_state=42
)
```

```python
LinearRegression_model = LinearRegression().fit(x_train_poly, y_train_poly)
```

```python
r2_score(y_test_poly, LinearRegression_model.predict(x_test_poly))
```

```python
poly_pipeline = PolynomialFeatures(degree=4)
LinearRegression_pipeline = LinearRegression()

Pipeline_model = make_pipeline(
    poly_pipeline,
    LinearRegression_pipeline
)

result = cross_val_score(Pipeline_model,
    x,
    y,
    cv=5,
    scoring="neg_mean_absolute_error"
)

result
```

```python
result.mean()
```

```python
mask = {"alpha": [0.001, 0.01, 0.05, 1, 2, 3, 7, 15, 30]}
```

```python
Lasso_model = Lasso()

mask_model = GridSearchCV(
    Lasso_model,
    mask,
    scoring="r2",
    cv=5
).fit(x_poly, y)

mask_model.best_params_
```

```python
mask = {"lasso__alpha": [0.001, 0.01, 0.05, 1, 2, 3, 7, 15, 30]}
```

```python
Pipeline_mask_model = make_pipeline(
    poly,
    Lasso_model,
)

result = GridSearchCV(Pipeline_mask_model, mask, scoring="r2", cv=5).fit(x_poly, y)
result.best_params_
```

## ЗАДАНИЕ 2

```python
df = pd.read_csv("files/data-3.csv")
df
```

```python
x, y = df[["x1", "x2"]].values, df["y"].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=.2,
    random_state=42
)
```

```python
mask = {"ridge__alpha": [0.001, 0.01, 0.05, 1, 2, 3, 7, 15, 30]}
```

```python
poly_model = PolynomialFeatures()
ridge_model = Ridge()

Pipeline_mask_model = make_pipeline(
    poly_model,
    ridge_model,
)

result = GridSearchCV(
    Pipeline_mask_model,
    mask,
    scoring="neg_mean_squared_error",
    cv=5
).fit(x_train, y_train)

result.best_params_
```

```python
mask = {"sgdregressor__alpha": [0.001, 0.01, 0.05, 1, 2, 3, 7, 15, 30]}
```

```python
poly_model = PolynomialFeatures()
scale_model = StandardScaler()

results = dict()
for penalty in ["l2", "l1", "elasticnet", None]:
    for loss in ["squared_error", "huber", "epsilon_insensitive", "squared_epsilon_insensitive"]:
        SGDRegressor_model = SGDRegressor(
            loss=loss,
            penalty=penalty
        )

        Pipeline_mask_model = make_pipeline(
            poly_model,
            scale_model,
            SGDRegressor_model,
        )

        result = GridSearchCV(
            Pipeline_mask_model,
            mask,
            scoring="neg_mean_squared_error",
            cv=6
        ).fit(x_train, y_train)

        results[f"{penalty}|{loss}"] = result.best_params_
```

```python
results
```

```python
mask = {"ridge__alpha": [0.001, 0.01, 0.05, 1, 2, 3, 7, 15, 30]}
```

```python
poly_model = PolynomialFeatures()

results = dict()
for solver in ["auto", "svd", "cholesky", "lsqr", "sparse_cg", "sag", "saga", "lbfgs"]:
    Ridge_model = Ridge(
        solver=solver,
        positive=True if solver in ["auto", "lbfgs"] else False
    )

    Pipeline_mask_model = make_pipeline(
        poly_model,
        Ridge_model,
    )

    result = GridSearchCV(
        Pipeline_mask_model,
        mask,
        scoring="neg_mean_squared_error",
        cv=6
    ).fit(x_train, y_train)

    results[f"{solver}"] = result.best_params_
```

```python
results
```

```python
mask = {"ridge__alpha": uniform(loc=0, scale=4)}
```

```python
poly_model = PolynomialFeatures()

results = dict()
for solver in ["auto", "svd", "cholesky", "lsqr", "sparse_cg", "sag", "saga", "lbfgs"]:
    Ridge_model = Ridge(
        solver=solver,
        positive=True if solver in ["auto", "lbfgs"] else False
    )

    Pipeline_mask_model = make_pipeline(
        poly_model,
        Ridge_model,
    )

    result = RandomizedSearchCV(
        Pipeline_mask_model,
        mask,
        scoring="neg_mean_squared_error",
        cv=6
    ).fit(x_train, y_train)

    results[f"{solver}"] = result.best_params_
```
