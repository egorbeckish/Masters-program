```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
x_data = np.array([1, 4])
y_data = np.array([2, 3])
const_m = np.array([[1, 0], [1, 1]])
```

```python
x_data_interpolation = np.linalg.inv(const_m) @ x_data
x_data_interpolation
```

```python
y_data_interpolation = np.linalg.inv(const_m) @ y_data
y_data_interpolation
```

```python
t = np.linspace(-10, 10, 100)
x = x_data_interpolation[0] + x_data_interpolation[1] * t
y = y_data_interpolation[0] + y_data_interpolation[1] * t
```

```python
plt.scatter(x_data, y_data, s=50, ec="black", zorder=3)
plt.plot(x, y, "r--")
plt.xlim((0, 5))
plt.ylim((-1, 5))
plt.annotate("Q", xy=(1, 2), xytext=(1, 2.3))
plt.annotate("R", xy=(4, 3), xytext=(4, 3.3))
plt.grid()
```
