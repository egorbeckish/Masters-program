```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x = np.linspace(0, 4 * np.pi)
x
```

```python
plt.plot(x, np.sin(x))
plt.grid()
```

```python
x = sp.symbols("x")
x
```

```python
f = x ** 2 + x - 10
f
```

```python
f.subs(x, 3)
```

```python
f2 = -x ** 2 + x + 10
f2
```

```python
f + f2
```

```python
f2.diff(x)
```

```python
sp.solve(f2, x)
```

```python
def f(x):
	return np.cos(3 * x - 3) ** 2
```

```python
x = np.linspace(0, 5, 1000)
y = f(x)
plt.plot(x, y)
plt.grid()
```

```python
def f(x):
	return np.where(
		(-2 <= x) & (x < 0),
		-x,
		np.where(
			(0 <= x)  & (x < 6),
			np.sin(x),
			None
		)
	)
```

```python
x = np.linspace(-3, 7, 100)
y = f(x)

plt.plot(x, y)
plt.grid()
plt.axvline(ls="--", c="r")
```
