```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

## Задача 1.1. Построить график функции 

```python
def y(x):
	return np.log10(x / (1 - x ** 2) ** .5)
```

```python
x = np.linspace(.01, 0.99, 100)
f = y(x)
f
```

```python
fig, ax = plt.subplots()
ax.plot(x, f)
ax.grid()
```

## Задача 1.2. Построить в одной системе координат графики функций 

```python
def y1(x):
	return x - np.cos(x)

def y2(x):
	return x - 2 * x ** 2
```

```python
x = np.linspace(-np.pi/2, np.pi, 100)
f1 = y1(x)
f2 = y2(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, f1, label=r"$y_1(x) = x - \cos(x)$")
ax.plot(x, f2, label=r"$y_2(x) = x - 2x^{2}$")
ax.grid()
ax.legend(fontsize=10, loc="lower left")
```

## Задача 1.3. В полярной системе координат построить график ρ 

```python
def r(theta):
	return 6 * np.cos(theta)
```

```python
theta = np.linspace(0, 2 * np.pi, 100)
f_r = r(theta)
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, f_r)
```
