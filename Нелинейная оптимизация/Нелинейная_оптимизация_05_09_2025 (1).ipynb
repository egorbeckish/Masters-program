```python
def f(x):
    result = np.zeros_like(x)
    mask1 = (-2 <= x) & (x < 0)
    mask2 = (0 <= x) & (x < 6)

    result[mask1] = -x[mask1]
    result[mask2] = np.sin(x[mask2])

    return result
```

```python
x = np.linspace(-2, 6, 10)
x
```

```python
y = f(x)
```

```python
x1 = np.linspace(-2, -0.001, 100)
x2 = np.linspace(0, 5.999, 200)

y1 = f(x1)
y2 = f(x2)
```

```python
plt.plot(x1, y1)
plt.plot(x2, y2)
```
